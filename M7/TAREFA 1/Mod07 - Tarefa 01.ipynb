#### 4) Qual variável é mais poderosa?

Considere as variáveis ```possui_email``` e ```posse_de_veiculo```. Faça uma tabela cruzada entre elas e responda qual delas te parece mais poderosa para prever a probabilidade de ```mau = 1```?

````python
#verificando quantos dados temos de cada variavel
tabela_cruzada = pd.crosstab(df['possui_email'], df['posse_de_veiculo'] ,margins=True, normalize=True)
tabela_cruzada
````

````python
pd.crosstab(df['possui_email'], df['posse_de_veiculo'], df['mau'], margins=True, normalize=True, aggfunc='mean')
````

#ANÁLISE 

A variavel possui_email verifica se a pessoa possui email ou não.
A variavel posse_de_veiculo verifica se a pessoa possui um veiculo ou não.
A variavel mau verifica se a pessoa é um bom pagador ou não.

Quando observamos a variavel posse_de_veiculo percebemos que não tem muito diferença na media do mau.
Quando olhamos para a variavel possui_email achamos uma media de mau maior para as pessoas que não possuem email. 

Assim concluimos que a variavel possui_email é mais poderosa
