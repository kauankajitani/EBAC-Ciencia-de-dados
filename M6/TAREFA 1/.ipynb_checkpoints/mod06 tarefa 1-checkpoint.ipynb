#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.

```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt

sinasc = pd.read_csv('SINASC_RO_2019_.csv')
sinasc.head()
```

```python
#verificacao de dados ausentes
sinasc[['ESTCIVMAE','GESTACAO', 'APGAR1', 'SEXO', 'APGAR5']].isna().sum().sort_values()
```

```python
#excluindo dados ausentes por serem poucos
sinasc = sinasc.dropna(subset = 'APGAR1')
sinasc = sinasc.dropna(subset = 'SEXO')
```

```python
#verificacao de dados ausentes
sinasc[['ESTCIVMAE','GESTACAO', 'APGAR1', 'SEXO', 'APGAR5']].isna().sum().sort_values()
```

```python
apgar1 = sns.countplot( x='APGAR1' , data = sinasc)
```

```python
apgar5 = sns.countplot( x='APGAR5' , data = sinasc)
```

```python
sexo = sns.countplot( x='SEXO' , data = sinasc)
```

```python
estcivmae = sns.countplot( x='ESTCIVMAE' , data = sinasc)
```

```python
gestacao = sns.countplot(x='GESTACAO' , data = sinasc)
gestacao.tick_params(axis = 'x',rotation = 30)
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.rc('figure', figsize=(10, 8))
fig, axes = plt.subplots(2,2)
plt.subplots_adjust(wspace=0.4, hspace=0.4)

ax = sns.countplot(ax = axes[0,0] ,x='GESTACAO' , data = sinasc)
ax.tick_params(axis = 'x',rotation = 30)

sns.countplot(ax = axes[0,1] , x='CONSULTAS' , data = sinasc)
sns.countplot(ax = axes[1,0] , x='SEXO' , data = sinasc)
sns.countplot(ax = axes[1,1] , x='ESTCIVMAE' , data = sinasc)
```
